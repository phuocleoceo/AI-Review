```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np


# def kmeans_display(X, label, n_cluster):
#     plot_color = ["b^", "go", "rs"]

#     print(">> Clusters : ")
#     Xi = []
#     for i in range(0, n_cluster):
#         Xi.append(X[label == i, :])
#         print(Xi[i])
#         plt.plot(Xi[i][:, 0], Xi[i][:, 1], plot_color[i], markersize=4, alpha=.8)

#     plt.axis('equal')
#     plt.show()


n_cluster = 3
X = np.array([[0,2,2],
              [2,2,0],
              [3,3,0],
              [3,2,2],
              [2,3,2],
              [2,3,1]])

kmeans = KMeans(n_clusters=n_cluster, random_state=0).fit(X)
print('Các center tìm được:')
print(kmeans.cluster_centers_)

pred_label = kmeans.predict(X)
# kmeans_display(X, pred_label, n_cluster)

# print("Kết quả phân nhóm : ")
# for i in range(0, n_cluster):
#     print(">> Nhóm",i+1,":",end="")
#     for xpl in X[pred_label == i, :]:
#         print(xpl, end=", ")
#     print("")
for i in range(kmeans.cluster_centers_.shape[0]):
   print(f">> Cụm thứ {i+1} : ")
   print(f"- Tâm : {kmeans.cluster_centers_[i]}")
   print("- Các điểm :",end="")
   for x in X[pred_label == i, :]:
      print(x, end=", ")
   print()
```
