```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('./input.csv')
data.head(20)
```

```python
X=data[["sepal_length","sepal_width","petal_length"	,"petal_width"]]
X = X.to_numpy().T
y=data["Labeling"]
# Encode nhãn cho biến dự đoán
label_encoder= LabelEncoder()
label_encoder.fit(y)
y = label_encoder.transform(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt


def prod(w, X):
    return np.dot(w.T, X)


def sigmoid(s):
    return 1 / (1 + np.exp(-s))


def my_logistic_sigmoid_regression(X, y, w_init, eta, epsilon=1e-3, M=10000):
    w = [w_init]
    N = X.shape[1]
    d = X.shape[0]
    count = 0
    check_w_after = 20
    while count < M:
        # mix data
        mix_id = np.random.permutation(N)
        for i in mix_id:
            xi = X[:, i].reshape(d, 1)
            yi = y[i]
            zi = sigmoid(np.dot(w[-1].T, xi))
            w_new = w[-1] + eta * (yi - zi) * xi
            count += 1
            # stopping criteria
            if count % check_w_after == 0:
                if np.linalg.norm(w_new - w[-check_w_after]) < epsilon:
                    return w
            w.append(w_new)
    return w
```

```python
Xbar = np.concatenate((np.ones((1, X.shape[1])), X), axis=0)
epsilon = 0.05
d = Xbar.shape[0]
w_init = np.random.randn(d, 1)
w = my_logistic_sigmoid_regression(Xbar, y, w_init, epsilon)
print(f">> Trọng số tìm được : ",end="")
print(w[-1].T)
```

```python
data_test = pd.read_csv('./output_04 .csv')
data_test.head(20)


X_test = data_test[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
X_test = X_test.to_numpy().T
Xbar_test = np.concatenate((np.ones((1, X_test.shape[1])), X_test), axis=0)
yi = sigmoid(np.dot(w[-1].T, Xbar_test))

def predict_flower(y_value):
    pred_value=None
    if y_value > 0.5:
        pred_value = 1
    else:
        pred_value = 0
    return label_encoder.classes_[pred_value]

pred_array = []
for y_value in yi[0]:
    pred_array.append(predict_flower(y_value))
print(pred_array)
```
